```python
# Scenario:
# You are a data science trainer preparing a demo for a group of first-time managers in e-commerce.
# You want to show them how to securely connect to Azure OpenAI from Google Colab, send a simple query,
#  and interpret the response. The demo emphasizes security (using secrets) and clarity (structured output).


from google.colab import userdata
from openai import OpenAI

# 🔐 Load API Key securely from Colab Secrets
api_key = userdata.get("OPENAI_API_KEY")

# ✅ Azure resource endpoint and deployment
endpoint = "https://chat-bot-001-resource.openai.azure.com/"
deployment_name = "gpt-4.1"

# Initialize client with proper Azure endpoint and API version
client = OpenAI(
    api_key=api_key,
    base_url=f"{endpoint}openai/deployments/{deployment_name}/",
    default_query={"api-version": "2024-02-01"}  # required for Azure
)

# Create a chat completion
completion = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {
            "role": "user",
            "content": "What is the capital of France?"
        }
    ]
)

# ✅ Access the message content correctly
print(completion.choices[0].message.content)
```
