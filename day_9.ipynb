```python
# Scenario
# A student is building a simple AI assistant in Google Colab for a class project. The assistant is
# hosted on Azure Foundry, and the student wants to ask it questions like:

# “What can you help with?”

# “Explain this topic simply.”

# “Give me a short summary.”

# Instead of talking directly to the model through a chat app, the student uses Python code to connect to the Azure service.

from google.colab import userdata

api_key = userdata.get("AZURE_OPENAI_API_KEY")
print("Loaded key:", api_key[:6] + "..." if api_key else "Missing")

from google.colab import userdata
from openai import OpenAI

my_endpoint = "https://20april.services.ai.azure.com/api/projects/proj-default"
my_agent = "18-april"
my_version = "2"

api_key = userdata.get("AZURE_OPENAI_API_KEY")

openai_client = OpenAI(
    base_url=f"{my_endpoint}/openai/v1",
    api_key=api_key
)

response = openai_client.responses.create(
    input=[{"role": "user", "content": "Tell me what you can help with."}],
    extra_body={
        "agent_reference": {
            "name": my_agent,
            "version": my_version,
            "type": "agent_reference"
        }
    },
)

print(response.output_text)
```

```python
from google.colab import userdata
from openai import OpenAI

# Config
my_endpoint = "https://18-april-chatbot.services.ai.azure.com/api/projects/proj-default"
my_agent = "18-april-chatbot"
my_version = "2"

api_key = userdata.get("AZURE_OPENAI_API_KEY_2")

client = OpenAI(
    base_url=f"{my_endpoint}/openai/v1",
    api_key=api_key
)

# Dynamic input
user_query = input("Ask your question: ")

# API call
response = client.responses.create(
    input=[
        {
            "role": "user",
            "content": user_query
        }
    ],

    extra_body={
        "agent_reference": {
            "name":my_agent,
            "version": my_version,
            "type": "agent_reference"
        }
    },
)

print("\n🤖 Bot:", response.output_text)
```
